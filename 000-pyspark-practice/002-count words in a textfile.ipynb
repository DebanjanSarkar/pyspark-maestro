```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").appName("Count words in a textfile").getOrCreate()
spark
```

```python
import pyspark.sql.functions as F
```

```python
paragraph_df = spark.read.format("text").load("./data/repeat_words.txt")
paragraph_df.show(truncate=False)
```

```python
words_df = (
    paragraph_df
    .withColumn("value", F.regexp_replace(F.col("value"), r"([.,])", "") )
    .withColumn("value", F.lower(F.col("value")))
    .withColumn("value", F.split("value", " "))
    .withColumn("value", F.explode(F.col("value")))
    .withColumnRenamed("value", "words")
)
```

```python
words_df.show()
```

```python
(
    words_df
    .groupBy("words")
    .agg( F.count(F.col("words")).alias("num_of_occurrence") )
    .orderBy( F.col("num_of_occurrence").desc() )
    .show()
)
```
